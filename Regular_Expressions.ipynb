**RegEx Module**

Python has a built-in package called re, which can be used to work with Regular Expressions.

Import the re module:


```python
import re

# When you have imported the re module, you can start using regular expressions:
```

```python
# Search the string to see if it starts with "The" and ends with "Spain":
txt = "The rain in train in Spain"
x = re.search("^The.*Spain$", txt)
print(x)
```

**RegEx Functions**

The re module offers a set of functions that allows us to search a string for a match:

The findall() function returns a list containing all matches.

The search() function searches the string for a match, and returns a Match object if there is a match. If there is more than one match, only the first occurrence of the match will be returned.

The split() function returns a list where the string has been split at each match.

The sub() function replaces the matches with the text of your choice.

The findall() function returns a list containing all matches.

```python
txt = "The rain in Spain"
x = re.findall("ai", txt)
print(x)

'''
The list contains the matches in the order they are found.

If no matches are found, an empty list is returned
'''
```

```python
# Return an empty list if no match was found:

txt = "The rain in Spain"
x = re.findall("Portugal", txt)
print(x)
```


The search() function searches the string for a match, and returns a Match object if there is a match. If there is more than one match, only the first occurrence of the match will be returned.

```python
txt = "Therain in Spain"
x = re.search("\s", txt)

print("The first white-space character is located in position:", x.start())
```

```python
# Make a search that returns no match:

txt = "The rain in Spain"
x = re.search("Portugal", txt)
print(x)
```

The split() function returns a list where the string has been split at each match:

```python
txt = "The rain in Spain"
x = re.split("\s", txt)
print(x)
```

```python
# Split at each white-space character:

txt = "The rain in Spain"
x = re.split("\s", txt, 1)
print(x)
```

```python
import re

# Example 1: Using \w
text = "Python123_ is great!"
result_w = re.findall(r'\w', text)  # Matches individual word characters
print("Matches for \\w:", result_w)
# Output: ['P', 'y', 't', 'h', 'o', 'n', '1', '2', '3', '_', 'i', 's', 'g', 'r', 'e', 'a', 't']

# Example 2: Using \W
result_W = re.findall(r'\W', text)  # Matches non-word characters
print("Matches for \\W:", result_W)
# Output: [' ', ' ', '!']

# Example 3: Splitting on Non-Word Characters (\W)
words = re.split(r'\W+', text)  # Split on non-word characters
print("Split text:", words)
# Output: ['Python123_', 'is', 'great', '']

# Example 4: Matching Words (\w+) and Non-Words (\W+)
result_words = re.findall(r'\w+', text)  # Matches sequences of word characters
print("Words:", result_words)
# Output: ['Python123_', 'is', 'great']

result_non_words = re.findall(r'\W+', text)  # Matches sequences of non-word characters
print("Non-Words:", result_non_words)
# Output: [' ', ' ', '!']
```

The sub() function replaces the matches with the text of your choice:

```python
# Replace every white-space character with the number 9:

txt = "The rain in Spain"
x = re.sub("\s", "9", txt)
print(x)
```

```python
# Replace the first 2 occurrences:

txt = "The rain in Spain"
x = re.sub("\s", "9", txt, 2)
print(x)
```

A Match Object is an object containing information about the search and the result.

Note: If there is no match, the value None will be returned, instead of the Match Object.

```python
# Do a search that will return a Match Object:

txt = "The rain in Spain"
x = re.search("ai", txt)
print(x) #this will print an object
```

The Match object has properties and methods used to retrieve information about the search, and the result:

* .span() returns a tuple containing the start-, and end positions of the match.

* .string returns the string passed into the function

* .group() returns the part of the string where there was a match

```python
# The regular expression looks for any words that starts with an upper case "S":

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.span())
```

```python
# Print the string passed into the function:

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.string)
```

```python
'''
Print the part of the string where there was a match.

The regular expression looks for any words that starts with an upper case "S":
'''

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.group())

# Note: If there is no match, the value None will be returned, instead of the Match Object.
```
