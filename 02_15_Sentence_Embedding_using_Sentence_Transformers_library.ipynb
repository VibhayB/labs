## Loading the SentenceTransformer Model

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
model.encode("I deposited money in the bank")
```

## Sentence List for Embedding

```python
sentences = [
    "I deposited money in the bank.",
    "We sat on the riverbank and watched the boats.",
    "The pilot had to bank the plane sharply to the left.",
    "The children sledded down the bank of the hill.",
    "The company has a large bank of customer data.",
    "There was a bank of monitors in the control room.",
    "We need to bank the fire with more logs.",
    "He made a bank shot to pocket the ball."
]
```

## Encoding Sentences to Embeddings

```python
embeddings = model.encode(sentences)
```

## Length of Embedding List

```python
len(embeddings)
```

## Accessing a Specific Embedding

```python
embeddings[7]
```

## Displaying Sentences and Their Embeddings

```python
for i, sentence in enumerate(sentences):
    print(f"Sentence: {sentence}")
    print(f"Embedding: {embeddings[i]}")
    print()
```
